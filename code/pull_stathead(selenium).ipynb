```python
options = webdriver.ChromeOptions()
options.add_argument('--disable-gpu')
options.add_experimental_option('extensionLoadTimeout', 60000)
driver = webdriver.Chrome(ChromeDriverManager().install(), options=options)

## get all team links
teams = set()

# Navigate to basketball-reference.com
driver.get("https://www.basketball-reference.com/leagues/NBA_2023.html")

src = driver.page_source
soup = BeautifulSoup(src)
# Find all the links
links = soup.find_all('a')

# Print the links that contain "/teams/" in the href attribute
for link in links:
    href = link.get('href')
    try:
        if ('/teams/' in href) and ('/2023' in href):
            result = re.search(r'/teams/(.*?)/2023', href)

            # Print the matched text
            if result:
                teams.add(result.group(1))
    except:
        pass
```

```python
## now get images
tm = list(teams)[0]
```

```python
filenames = next(walk('/Users/meganmuniz/DGMD/TeamImages'), (None, None, []))[2]
current = set()
for pic in filenames:
    current.add(pic[:-4])
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--disable-gpu')
options.add_experimental_option('extensionLoadTimeout', 60000)
driver = webdriver.Chrome(ChromeDriverManager().install(), options=options)


for tm in set(teams)-current:
    try:
        driver.get(f'https://www.basketball-reference.com/teams/{tm}')
        src = driver.page_source
        soup = BeautifulSoup(src)

        # Find the image element
        img = soup.find('img', class_='teamlogo')

        # Get the image URL
        img_url = img.get('src')

        # Download the image
        response = requests.get(img_url, stream=True)

        # Save the image to a file
        with open(f'TeamImages/{tm}.png', 'wb') as f:
            f.write(response.content)
            f.close()
    except:
        pass
```
